```python
df_confirmed=pd.read_csv('../data/processed/COVID_small_flat_table_confirmed.csv',sep=';')
df_confirmed.sort_values('date',ascending=True).head()
```

```python
df_recovered=pd.read_csv('../data/processed/COVID_small_flat_table_recovered.csv',sep=';')  
```

```python
df_deaths=pd.read_csv('../data/processed/COVID_small_flat_table_deaths.csv',sep=';')  
```

```python
country_list=df_confirmed.columns[1:]
country_dropdown=[]
for country in country_list:
    country_dropdown.append({'label':country,'value':country})
```

```python
import dash
dash.__version__
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()

fig1= go.Figure()
fig2= go.Figure()
fig3= go.Figure()
fig= go.Figure()
fig_recover_death=go.Figure()

app.layout = html.Div([

    html.Div([
    
    html.Label('Multi-Select Country'),
    
    dcc.Dropdown(
        id='country_drop_down',
        options=country_dropdown,
        value=['US', 'Germany'], # which are pre-selected
        multi=True
    ),
    
    dcc.RadioItems(
                id='yaxis-type',
                options=[{'label': i, 'value': i} for i in ['Linear', 'Log']],
                value='Log',
                labelStyle={'display': 'inline-block'}),
        
    dcc.Graph(figure=fig, id='main_window_slope')
    ]),
    
    html.Div([
    
    html.Label('Recovery and death cases'),
    
    dcc.Dropdown(
        id='single_input',
        options=country_dropdown,
        value='US', # which are pre-selected
        
    ),    
        
    dcc.Graph(figure=fig_recover_death, id='recover_death')
    ]),
    
  html.Div([
    
#     html.Label('Recovery trend for each country'),    
 
    dcc.Graph(figure=fig1, id='newly_confirmed')
    ]),
    
     html.Div([
    
#     html.Label('Recovery trend for each country'),    
 
    dcc.Graph(figure=fig2, id='newly_deaths')
    ]),
     html.Div([
    
#     html.Label('Recovery trend for each country'),    
 
    dcc.Graph(figure=fig3, id='newly_recovered')
    ]),   
    

])




```

```python
from dash.dependencies import Input, Output

@app.callback(
    Output('main_window_slope', 'figure'),
    [Input('country_drop_down', 'value'),
     Input('yaxis-type', 'value')])
def update_figure(country_list, yaxis_type):
    
    traces = [] 
    for each in country_list:
        traces.append(dict(x=df_confirmed.date,
                                y=df_confirmed[each],
                                mode='markers+lines',
                                opacity=0.9,
                                line_width=2,
                                marker_size=4, 
                                name=each
                        )
                )
        
    return {
            'data': traces,
            'layout': dict (
                width=1280,
                height=720,
                plot_bgcolor='rgba(1,1,1,1)',                
#                 xaxis_title="Time",
#                 yaxis_title="Confirmed infected people (source johns hopkins csse, log-scale)",
                xaxis={'tickangle':-45,
                        'nticks':20,
                        'tickfont':dict(size=14,color="#7f7f7f"),
                       "title": "Time"  
                        
                      },
                yaxis={'type':'Linear' if yaxis_type == 'Linear' else 'log',
                       'range':'[1.1,5.5]',
                       "title": "Confirmed infected people (source johns hopkins csse, log-scale)"
                      }
        )
    }
```

```python
@app.callback(
    Output('recover_death', 'figure'),
    [Input('single_input', 'value')])
def update_figure(country_name):
    
    traces = [] 
    traces.append(dict(x=df_recovered[30:].date,
                                y=df_recovered[30:][country_name],
                                mode='markers+lines',
                                opacity=0.9,
                                line_width=2,
                                marker_size=4, 
                                name='Recovered cases'
                        )),
    traces.append(dict(x=df_deaths[30:].date,
                                y=df_deaths[30:][country_name],
                                mode='markers+lines',
                                opacity=0.9,
                                line_width=2,
                                marker_size=4, 
                                name='Death cases'
                        )
                )
        
    return {
            'data': traces,
            'layout': dict (
                width=1280,
                height=720,
                plot_bgcolor='rgba(1,1,1,1)',
               title = "Stats for country " + country_name,
                xaxis={'tickangle':-45,
                        'nticks':20,
                        'tickfont':dict(size=14,color="#7f7f7f"),
                       'title' : 'Time'
                        
                      },
                yaxis={'type':'log',
                       'range':'[1.1,5.5]',
                       'title' : 'Comparison between Recovery rate vs. Death rate.'
                      }
        )
    }
```

```python
@app.callback(
    Output('newly_confirmed', 'figure'),
    [Input('single_input', 'value')])
def update_figure(list_of_countries):
    
    traces = [] 
    traces.append(dict(x=df_confirmed.date[40:],
                                y=df_confirmed[list_of_countries].diff()[40:],
                                type = "bar",
                                orientation = 'v'
                        )
                )
        
    return {
            'data': traces,
            'layout': dict (
                width=1280,
                height=720,
                plot_bgcolor='rgba(1,1,1,1)',
                title = "Stats for country " + list_of_countries,
                xaxis = {'title' : "Time"                  
                        
                      },
                yaxis = {'title' : "Newly confirmed cases"                  
                        
                      },
#                 xaxis_title="Time",
#                 yaxis_title="Recovery rate (source johns hopkins csse, log-scale)",
     
        )
    }
```

```python
@app.callback(
    Output('newly_deaths', 'figure'),
    [Input('single_input', 'value')])
def update_figure(list_of_countries):
    
    traces = [] 
    traces.append(dict(x=df_deaths.date[30:],
                                y=df_deaths[list_of_countries].diff()[30:],
                                type = "bar",
                                orientation = 'v'
                        )
                )
        
    return {
            'data': traces,
            'layout': dict (
                width=1280,
                height=720,
                plot_bgcolor='rgba(1,1,1,1)',
                title = "Stats for country " + list_of_countries,
                  xaxis = {'title' : "Time"                  
                        
                      },
                yaxis = {'title' : "Newly death cases"                  
                        
                      },
#                 xaxis_title="Time",
#                 yaxis_title="Recovery rate (source johns hopkins csse, log-scale)",
     
        )
    }
```

```python
@app.callback(
    Output('newly_recovered', 'figure'),
    [Input('single_input', 'value')])
def update_figure(list_of_countries):
    
    traces = [] 
    traces.append(dict(x=df_recovered.date[30:],
                                y=df_recovered[list_of_countries].diff()[30:],
                                type = "bar",
                                orientation = 'v'
                        )
                )
        
    return {
            'data': traces,
            'layout': dict (
                width=1280,
                height=720,
                plot_bgcolor='rgba(1,1,1,1)',
                title = "Stats for country " + list_of_countries,
                xaxis = {'title' : "Time"                  
                        
                      },
                yaxis = {'title' : "Newly recovered cases"                  
                        
                      },
#                 xaxis_title="Time",
#                 yaxis_title="Recovery rate (source johns hopkins csse, log-scale)",
     
        )
    }
```
